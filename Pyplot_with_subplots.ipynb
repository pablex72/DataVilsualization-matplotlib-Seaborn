```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.show()
```

```python
#plt.plot(x,y,'colorsimbolo')-->plot graphic de lineas
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'rs-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

BOXPLOT distribucion de datos

```python
plt.boxplot(x)
plt.show()
```

Appliying Subplot to show more graphics

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
