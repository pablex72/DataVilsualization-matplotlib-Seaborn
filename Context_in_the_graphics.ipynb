```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
####METODO ORIENTADO A OBJETOS
fig, axes = plt.subplots(1,2,figsize = (5,5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title('Relacion X -Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y -X')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es mi titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', bbox_to_anchor=(0.5,0.2))
```

Modificaciones tipo MATLAB

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'y-')
ax.plot(x,x+3,'b.')
ax.plot(x,x+4,'g:')
```

PYPLOT

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.plot(x,x+1,color='green', alpha=0.1, linewidth=5, linestyle='-',marker='o',markersize=8,markerfacecolor='#ffffff')
ax.plot(x,x+2,color='blue', linewidth=8, linestyle='--',marker='x')
ax.plot(x,x+3,color='#aaaaaa',marker='v')
ax.plot(x,x+4,color='#bbbbbb',marker='o')
```
